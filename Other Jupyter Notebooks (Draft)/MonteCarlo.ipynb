```python
import numpy as np
sims = 100000000

A = np.random.uniform(1, 5, sims)
B = np.random.uniform(2, 6, sims)

duration = A + B
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (5, 2))
plt.hist(duration, density = True)
#plt.axvline(9, color = 'r')
plt.show()
print((duration > 9).sum()/sims)
```
